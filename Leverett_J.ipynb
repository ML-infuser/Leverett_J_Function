```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('S1Data.XLSX')
df
```

```python
x = df.iloc[10:,1:5]
x
```

```python
plt.scatter(x['Unnamed: 3'],x['Unnamed: 4'],color = 'r')
plt.plot(x['Unnamed: 3'],x['Unnamed: 4'])
```

```python
from sklearn.linear_model import LinearRegression as lr
lr1 = lr()
xtrain = np.array(x['Unnamed: 3']).reshape(-1,1)
ytrain = np.array(x['Unnamed: 4']).reshape(-1,1)
```

```python
lr1.fit(xtrain,ytrain)
```

```python
xpred = np.array([50,55,60]).reshape(-1,1)
```

```python
ypred = lr1.predict(xpred)
ypred
```
